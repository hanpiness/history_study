### 方法1：标准化缩放法
- 对简单示例二维矩阵的列数据进行

```python
from sklearn import preprocessing
import numpy as np
```

#### 方法1: 标准化缩放法 对简单实例二维矩阵的列数据进行

```python
X = np.array([[0,0],# 采用numpy的array表示，因为要用到其mean等函数，而list没有这些函数
            [0,0],
            [100,1],
            [1,1]])
X_mean = X.mean(axis = 0)
X_std = X.std(axis = 0)
# print(X_std)
# print(X_mean)
X1 = (X-X_mean)/X_std
print(X1)
# 利用一个集成的方法
# X2 = preprocessing.scale(X)
# print(X2)
```

#### 方法1： 标准化缩放法 对 iris数据二维矩阵的列数据进行

```python
from sklearn import datasets
iris = datasets.load_iris()
X_scale = preprocessing.scale(iris.data)
print(X_scale)
```

#### 方法2: 区间缩放法 对简单示例二维矩阵的列数据进行

```python
from sklearn.preprocessing import MinMaxScaler
data = [[0,0],
       [0,0],
       [100,1],
       [1,1]]
scaler = MinMaxScaler()
print(scaler.fit(data))
print(scaler.transform(data)) #用此方法，必须使用fit先进行预处理计算均值和标准差
```

#### 方法2： 区间缩放法 对iris数据二维矩阵的列数据

```python
from sklearn.preprocessing import MinMaxScaler
data = iris.data
scaler = MinMaxScaler()
print(scaler.fit(data))
print(scaler.transform(data))
```
