```python
# 图像标准化
def normalize_image(img):
    img = 255*(img + 1.0)/2.0
    return tf.cast(img,tf.uint8) # 类型转换
```

```python
# 图像可视化
def show_image(img):
    display.display(PIL.Image.fromarray(np.array(img)))  # 图像显示
```

```python
# 保存图像文件
def save_image(img,file_name):
    PIL.Image.fromarray(np.array(img)).save(file_name)
```

```python
# 定义图像噪声
img_noise = np.random.uniform(size = (300,300,3)) + 100.0
img_noise = img_noise.astype(np.float32)   # dtype转换为float32
show_image(normalize_image(img_noise))
```
